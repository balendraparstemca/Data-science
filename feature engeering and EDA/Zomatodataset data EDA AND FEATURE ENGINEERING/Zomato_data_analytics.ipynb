````python
df_country=pd.read_excel('Country-Code.xlsx')
df_country.head()
````

````python
finaldf=pd.merge(df,df_country,on="Country Code",how='left')
finaldf.head()
````

````python
finaldf.dtypes
````

````python
finaldf.columns
````

````python
finaldf.Country.value_counts()
````

````python
countryname=finaldf.Country.value_counts().index
````

````python
country_val=finaldf.Country.value_counts().values
````

````python
plt.pie(country_val[:3],labels=countryname[:3],autopct='%1.2f%%')
````

````python
ratings=finaldf.groupby(['Aggregate rating','Rating color','Rating text']).size()
ratings
````

````python
ratings.reset_index()
````

````python
ratings=ratings.reset_index().rename(columns={0:'Rating count'})
ratings
````



 ```
# Observation
When Rating is between 4.5 to 4.9---> Excellent
When Rating are between 4.0 to 3.4--->very good
when Rating is between 3.5 to 3.9----> good
when Rating is between 3.0 to 3.4----> average
when Rating is between 2.5 to 2.9----> average
when Rating is between 2.0 to 2.4----> Poor
```
