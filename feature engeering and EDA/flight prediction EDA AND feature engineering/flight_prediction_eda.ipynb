```python

finaldf['Date']=finaldf['Date_of_Journey'].str.split('/').str[0]
finaldf['Month']=finaldf['Date_of_Journey'].str.split('/').str[1]
finaldf['Year']=finaldf['Date_of_Journey'].str.split('/').str[2]

```

```python
finaldf.head()
```

```python
finaldf['Date']=finaldf['Date'].astype(int)
finaldf['Month']=finaldf['Month'].astype(int)
finaldf['Year']=finaldf['Year'].astype(int)
```

```python
finaldf.drop('Date_of_Journey',axis=1,inplace=True)
```

```python
finaldf.head()
```

```python
finaldf.info()
```

```python
finaldf['Arrival_Time']=finaldf['Arrival_Time'].str.split(' ').str[0]
finaldf.head()
```

```python
finaldf['Arrival_hour']=finaldf['Arrival_Time'].str.split(':').str[0]
finaldf['Arrival_min']=finaldf['Arrival_Time'].str.split(':').str[1]
finaldf.head()
```

```python
finaldf['Arrival_hour']=finaldf['Arrival_hour'].astype(int)
finaldf['Arrival_min']=finaldf['Arrival_hour'].astype(int)
finaldf.info()
```

```python
#final_df.drop('Arrival_Time',axis=1,inplace=True)
finaldf.head()
```

```python
finaldf['Dept_hour']=finaldf['Dep_Time'].str.split(':').str[0]
finaldf['Dept_min']=finaldf['Dep_Time'].str.split(':').str[1]
finaldf['Dept_hour']=finaldf['Dept_hour'].astype(int)
finaldf['Dept_min']=finaldf['Dept_min'].astype(int)
finaldf.drop('Dep_Time',axis=1,inplace=True)
```
